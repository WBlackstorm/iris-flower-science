```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('IRIS.csv')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['species'].unique()
```

```python
plt.boxplot(df['sepal_length'])
plt.title("Sepal Length", loc="center", fontsize=12)
plt.show()
```

```python
plt.boxplot(df['sepal_width'])
plt.title("Sepal Width", loc="center", fontsize=12)
plt.show()
```

```python
# possible outliers in sepal width
df.loc[(df['sepal_width'] > 4) | (df['sepal_width'] < 2.2)]
```

```python
plt.boxplot(df['petal_length'])
plt.title("Petal Length", loc="center", fontsize=12)
plt.show()
```

```python
plt.boxplot(df['petal_width'])
plt.title("Petal Width", loc="center", fontsize=12)
plt.show()
```

```python
setosa = df.loc[df['species'] == 'Iris-setosa']
virginica = df.loc[df['species'] == 'Iris-virginica']
versicolor = df.loc[df['species'] == 'Iris-versicolor']

fig, ax = plt.subplots()

ax.scatter(setosa['sepal_length'], setosa['sepal_width'], label='Setosa')
ax.scatter(virginica['sepal_length'], virginica['sepal_width'], label='Virginica')
ax.scatter(versicolor['sepal_length'], versicolor['sepal_width'], label='Versicolor')
ax.legend()

plt.title("Sepal Dispersion", loc="center", fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(setosa['petal_length'], setosa['petal_width'], label='Setosa')
ax.scatter(virginica['petal_length'], virginica['petal_width'], label='Virginica')
ax.scatter(versicolor['petal_length'], versicolor['petal_width'], label='Versicolor')
ax.legend()

plt.title("Petal Dispersion", loc="center", fontsize=12)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(setosa['sepal_width'], setosa['petal_width'], label='Setosa')
ax.scatter(virginica['sepal_width'], virginica['petal_width'], label='Virginica')
ax.scatter(versicolor['sepal_width'], versicolor['petal_width'], label='Versicolor')
ax.legend()

plt.title("Sepal/Petal Width Dispersion", loc="center", fontsize=12)
plt.show()
```
